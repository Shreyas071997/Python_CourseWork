# Section 1: Control Structures (16 marks)


Explanation and Documentation:
This code demonstrates how to load and inspect a dataset using the Pandas library in Python. 
It begins by importing Pandas and specifying the file path to the CSV file, making the file path a raw string to handle drawbacks correctly.
The dataset is then read using pd.read_csv() into a Pandas DataFrame called athlete_data, which organizes the data into a structured, tabular format.
The head() method is used to preview the first five rows, providing a quick look at the dataset to check if it is loaded correctly.
Finally, the info() method summarizes the dataset, showing the number of rows and columns, column names, data types, non-null counts, and memory usage.
This process ensures the dataset is ready for further analysis, to identify data quality issues, exploring trends, and performing visualizations.

```python
import pandas as pd
file_path = r"C:\Users\Shreyas\Desktop\Athlete_injury_dataset.csv"
athlete_data = pd.read_csv(file_path)
print(athlete_data.head())
print(athlete_data.info())
```

Explanation and Documentation of Control Structures:
Control structures is used to filter and analyze specific subsections of data from the dataset. 
This code uses for loops with if-else control structures to analyze athlete data manually. 
The first block goes through each row of the dataset to check if the Fatigue_Score is greater than 70. If the condition is met, the athletes ID
and fatigue score are added to the high_fatigue_list, which is printed if it contains any entries, else a message indicates no athletes meet the
criteria. The second block uses a dictionary to count the number of athletes in each Position. For every position in the dataset, the code checks 
if the position already exists in the dictionary, if so it increases the count, otherwise initializes it. 
This approach highlights the use of conditional logic to filter and categorize data ensuring flexibility in manual data processing.

```python
high_fatigue_list = []
for index, row in athlete_data.iterrows():
    if row['Fatigue_Score'] > 70:
        high_fatigue_list.append({'Athlete_ID': row['Athlete_ID'], 'Fatigue_Score': row['Fatigue_Score']})
    else:
        continue  

if high_fatigue_list:
    print("Athletes with Fatigue Score above 70:")
    for athlete in high_fatigue_list:
        print(athlete)
else:
    print("No athletes with Fatigue Score above 70.")


position_counts = {}
for position in athlete_data['Position']:
    if position in position_counts:
        position_counts[position] += 1
    else:
        position_counts[position] = 1

print("\nAthletes by Position:")
for position, count in position_counts.items():
    print(f"{position}: {count}")
```

Reflection on results:
The control structures reveal that no athletes in the dataset have a Fatigue_Score above 70, as indicated by an empty DataFrame.
The value_counts() function shows that the dataset consists of 70 forwards, 66 centers, and 64 guards, highlighting the distribution
of athletes by position.

# Section 2: Functions and Modules (16 marks)  

Explanation and Documentation of Functions and Modules:
This code defines two modular functions to analyze specific aspects of the athlete dataset using the Pandas library.
The average_performance_by_position(position) function calculates the average performance score for athletes in a specified position by filtering
the dataset and using the mean() method, giving insights into positional performance trends. The high_acl_risk(threshold) function identifies
athletes with ACL Risk Scores exceeding a given threshold by filtering the dataset and returning only the relevant Athlete_ID and ACL_Risk_Score
columns. These functions are designed to be reusable and focus on targeted data analysis and efficient filtering. 

```python
def average_performance_by_position(position):
    data = athlete_data[athlete_data['Position'] == position]
    return data['Performance_Score'].mean()


print("Average performance for Center:", average_performance_by_position('Center'))


def high_acl_risk(threshold):
    return athlete_data[athlete_data['ACL_Risk_Score'] > threshold][['Athlete_ID', 'ACL_Risk_Score']]


print("Athletes with ACL Risk Score above 60:\n", high_acl_risk(60))
```

Reflection on results:
The results show that the average performance score for athletes in the "Center" position is approximately 72.65, indicating consistent performance 
levels for this role. The high_acl_risk function identified multiple athletes with ACL Risk Scores above 60, highlighting individuals
who may require monitoring or preventative measures to mitigate injury risk.

# Section 3: Data Handling with Pandas (16 marks)  

Explanation and Documentation of Data Handling and Pandas:
This code demonstrates data handling using Pandas by cleaning, grouping, and sorting operations to extract meaningful insights. 
Missing values in the dataset are removed using dropna(), ensuring data integrity for subsequent analysis. 
The dataset is then grouped by the Position column to calculate the average Fatigue_Score for each position using groupby() and mean(),
providing insights into fatigue levels across different roles. The dataset is then sorted in descending order by Performance_Score using 
sort_values(), and the top 10 performers are displayed by selecting relevant columns. These Pandas operations enable efficient data cleaning and
extraction.

```python
athlete_data = athlete_data.dropna()  
print("Dataset after handling missing values:\n", athlete_data.info())

average_fatigue_by_position = athlete_data.groupby('Position')['Fatigue_Score'].mean()
print("Average Fatigue Score by Position:\n", average_fatigue_by_position)


top_performers = athlete_data.sort_values(by='Performance_Score', ascending=False)
print("Top 10 Performers:\n", top_performers[['Athlete_ID', 'Performance_Score']].head(10))
```

Reflection on results:
The results confirm the dataset is clean, with all 200 entries having no missing values after handling.
The output shows that guards have the highest average fatigue score (5.44), followed by centers (4.70) and forwards (4.66), indicating 
position-specific fatigue patterns. The top 10 performers, with scores ranging from 98 to 99, demonstrate Pandas to sort and rank data

# Section 4: Data Visualization (16 marks)  

Explanation and Documentation of Data Visualization: 
Seaborn and Matplotlib is used to create visualizations that provide insights into athlete data. 
The visualize_injury_trends function generates a bar chart showing the number of injured versus non-injured athletes, helping to identify injury
patterns. visualize_fatigue_levels function creates a histogram with a KDE overlay to display the distribution of fatigue scores, revealing
trends and potential outliers in fatigue levels. The visualize_performance_scores function produces a line plot of performance scores sorted 
in descending order, highlighting top-performing athletes. These visualizations leverage Pandas for data processing and Seaborn for aesthetics, making
it easy to interpret key metrics like injuries, fatigue, and performance in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"C:\Users\Shreyas\Desktop\Athlete_injury_dataset.csv"
data = pd.read_csv(file_path)


sns.set(style="whitegrid")


def visualize_injury_trends(data):
    injury_counts = data['Injury_Indicator'].value_counts().reset_index()
    injury_counts.columns = ['Injury_Status', 'Count']
    injury_counts['Injury_Status'] = ['Non-Injured', 'Injured']  
    
    plt.figure(figsize=(8, 5))
    sns.barplot(data=injury_counts, x='Injury_Status', y='Count', hue='Injury_Status', dodge=False,
                palette=sns.color_palette("Blues", len(injury_counts)))
    plt.title("Injury Trends: Injured vs Non-Injured Athletes")
    plt.ylabel("Number of Athletes")
    plt.xlabel("Injury Status")
    plt.legend([], [], frameon=False)  
    plt.show()


def visualize_fatigue_levels(data):
    plt.figure(figsize=(8, 5))
    sns.histplot(data['Fatigue_Score'], kde=True, color="orange", bins=15)
    plt.title("Fatigue Levels Distribution")
    plt.xlabel("Fatigue Score")
    plt.ylabel("Frequency")
    plt.show()


def visualize_performance_scores(data):
    plt.figure(figsize=(10, 6))
    sorted_data = data.sort_values(by='Performance_Score', ascending=False)
    plt.plot(sorted_data['Performance_Score'].values, marker='o', linestyle='-', color="green")
    plt.title("Performance Scores Across Athletes")
    plt.xlabel("Athlete Index (Sorted by Performance)")
    plt.ylabel("Performance Score")
    plt.grid(True)
    plt.show()


visualize_injury_trends(data)
visualize_fatigue_levels(data)
visualize_performance_scores(data)
```

```python
Reflection on results 
The visualizations provide valuable insights into the athlete dataset.
The injury trends chart reveals that most athletes are non-injured, with a small proportion suffering from injuries, highlighting the datasets 
overall low injury rate. Fatigue levels distribution shows a relatively even spread of fatigue scores, with no extreme outliers, suggesting balanced
workload management across athletes. 
The performance scores visualization predicts a gradual decline from top-performing athletes to lower performers, clearly highlighting the distribution
of excellence within the group. These visualizations effectively convey trends and distributions for informed analysis.
```
